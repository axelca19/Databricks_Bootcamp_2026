# Loop Logic

```sql
%python

# Loop through the configuration list
for item in INGESTION_CONFIG:
    print(f"üöÄ Ingesting: {item['path']}...")
    
    try:
        # 1. Read the CSV (Using the path from config)
        df = (spark.read
              .format("csv")
              .option("header", True)
              .option("inferSchema", True)
              .load(item['path']))
        
        # 2. Write to Delta Table (Using the table name from config)
        (df.write
           .format("delta")
           .mode("overwrite")
           .saveAsTable(item['table']))
           
        print(f"‚úÖ Created Table: {item['table']}")
        
    except Exception as e:
        print(f"‚ùå Failed to load {item['path']}: {e}")
```
